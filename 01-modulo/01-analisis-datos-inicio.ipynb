## Realice el análisis de correlación y covarianza entre los atributos de precios de soja y maíz.

```python
%reset -f
```

```python
import pandas as pd
```

```python
archivo3 = 'soya_maiz2.xlsx'
df3 = pd.read_excel(archivo3)
df3.set_index(df3.columns[0], inplace=True)
df3['Fecha'] = pd.to_datetime(df3['Fecha'])
df3.head()
```

```python
correlacion = df3['Precios - Maíz'].corr(df3['Precios - Soja'])
print(correlacion)
```

```python
covarianza = df3['Precios - Maíz'].cov(df3['Precios - Soja'])
print(covarianza)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que tienes un DataFrame llamado df con columnas 'Precios - Soja' y 'Precios - Maíz'

# Graficar el gráfico de dispersión
plt.figure(figsize=(8, 6))
plt.scatter(df3['Precios - Soja'], df3['Precios - Maíz'], alpha=0.5, color='blue')
plt.title('Gráfico de Dispersión - Precios de Soja vs Maíz')
plt.xlabel('Precios - Soja')
plt.ylabel('Precios - Maíz')
plt.grid(True)
plt.show()
```

Conclusión: Los conjuntos de datos muestran una correlación positiva bastante fuerte (0.749) y una covarianza positiva (15.49), lo que indica una relación lineal entre las variables y una tendencia a variar en la misma dirección.
