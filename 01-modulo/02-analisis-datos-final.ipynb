## 3. Realice el análisis de correlación y covarianza entre los atributos de precios de soja y maíz.

```python
import pandas as pd

# Suponiendo que 'df' contiene tus datos

# Calcular la matriz de correlación
correlacion = df[['Precios - Soja', 'Precios - Maíz']].corr()

# Calcular la matriz de covarianza
covarianza = df[['Precios - Soja', 'Precios - Maíz']].cov()

# Mostrar los resultados de correlación y covarianza
print("Matriz de Correlación:")
print(correlacion)

print("\nMatriz de Covarianza:")
print(covarianza)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Suponiendo que tienes un DataFrame llamado df con columnas 'Precios - Soja' y 'Precios - Maíz'

# Calcular la correlación
correlation = df['Precios - Soja'].corr(df['Precios - Maíz'])

# Crear el gráfico de dispersión con línea de regresión
plt.figure(figsize=(8, 6))
sns.regplot(x='Precios - Soja', y='Precios - Maíz', data=df, scatter_kws={'color': 'blue', 'alpha': 0.5},
            line_kws={'color': 'red', 'alpha': 0.7})
plt.title('Gráfico de Dispersión con Línea de Regresión')
plt.xlabel('Precios - Soja')
plt.ylabel('Precios - Maíz')

# Mostrar información sobre la correlación en el gráfico
plt.text(0.1, 0.9, f'Correlación: {correlation:.2f}', horizontalalignment='left',
         verticalalignment='center', transform=plt.gca().transAxes)
plt.grid(True)
plt.show()
```
